```python
import pandas as pd
import os

def load_csv_file(path,filename):
    csv_path=os.path.join(path,filename)
    return pd.read_csv(csv_path)
```

```python
data=load_csv_file('./','electricity_dataset.csv')
data.head()
```

```python
data.info()
```

```python
data['ForecastWindProduction']=pd.to_numeric(data['ForecastWindProduction'],errors='coerce')
data['SystemLoadEA']=pd.to_numeric(data['SystemLoadEA'],errors='coerce')
data['SMPEA']=pd.to_numeric(data['SMPEA'],errors='coerce')
data['ORKTemperature']=pd.to_numeric(data['ORKTemperature'],errors='coerce')
data['ORKWindspeed']=pd.to_numeric(data['ORKWindspeed'],errors='coerce')
data['CO2Intensity']=pd.to_numeric(data['CO2Intensity'],errors='coerce')
data['ActualWindProduction']=pd.to_numeric(data['ActualWindProduction'],errors='coerce')
data['SystemLoadEP2']=pd.to_numeric(data['SystemLoadEP2'],errors='coerce')
data['SMPEP2']=pd.to_numeric(data['SMPEP2'],errors='coerce')
data['HolidayFlag']=pd.to_numeric(data['HolidayFlag'],errors='coerce')
data['DayOfWeek']=pd.to_numeric(data['DayOfWeek'],errors='coerce')
data['WeekOfYear']=pd.to_numeric(data['WeekOfYear'],errors='coerce')
data['Day']=pd.to_numeric(data['Day'],errors='coerce')
data['Month']=pd.to_numeric(data['Month'],errors='coerce')
data['Year']=pd.to_numeric(data['Year'],errors='coerce')
data['PeriodOfDay']=pd.to_numeric(data['PeriodOfDay'],errors='coerce')
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
data.hist(bins=50,figsize=(20,15))
```

```python
import sklearn.model_selection as ms
data_train,data_test=ms.train_test_split(data,test_size=0.2,random_state=42)
```

```python
data_train.plot(kind='scatter',x='ORKWindspeed',y='ORKTemperature',alpha=0.4,
          s=data_train['SystemLoadEA']/100,label='System Load',figsize=(10,7),
          c='SMPEA',cmap=plt.get_cmap('jet'),colorbar=True)
```

```python
#Data correlation with Label[SMPEP2]
data_train=data_train.drop(columns=['DateTime','Holiday'],axis=1)
data_correlation=data_train.corr()
data_correlation['SMPEP2'].sort_values(ascending=False)
```

```python
import seaborn as sns

data_correlation=data_train.corr(method='pearson')
plt.figure(figsize=(16,12))
sns.heatmap(data_correlation,cmap='coolwarm',annot=True)
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(data_train)
```

```python
data_train.median().values
```

```python
X=imputer.transform(data_train)
```

```python
data_train=pd.DataFrame(X,columns=data_train.columns)
```

```python
data_train_label=data_train['SMPEP2']
data_train=data_train.drop(columns='SMPEP2',axis=1)
```

```python
data_train.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
num_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scaler',StandardScaler())
])
data_train=num_pipeline.fit_transform(data_train)
```

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
linear_model.fit(data_train,data_train_label)
```

```python
data_test.head()
```

```python
data_test_label=data_test['SMPEP2']
data_test=data_test.drop(columns=['DateTime','Holiday','SMPEP2'],axis=1)
prepared_testing_data=num_pipeline.fit_transform(data_test)
```

```python
print(prepared_testing_data)
```

```python
def predict_prices():
    
    return linear_model.predict(prepared_testing_data[:5])
```

```python
predict_prices()
```

```python
print('labels',list(data_test_label[:5]))
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error
def root_mean_error(label,modzel):
    model_reg_mse=mean_absolute_error(label,modzel)
    model_rsme=np.sqrt(model_reg_mse)
    return model_rsme
root_mean_error(data_train_label,linear_model.predict(data_train))
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_model=DecisionTreeRegressor()
tree_model.fit(data_train,data_train_label)
```

```python
tree_model.predict(prepared_testing_data[:5])
```

```python
#Calculating root mean squared error of the Decision regressor 
root_mean_error(data_train_label,tree_model.predict(data_train))
```
